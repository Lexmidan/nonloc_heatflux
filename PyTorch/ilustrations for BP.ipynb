```python
initiate=pd.read_csv(f'.\\Data\\results\\30-08-2023, 12-22-11_heatflux_300_Sigmoid0.csv', index_col=0)
profiles_sigmoid=initiate[initiate.columns[0]].values

for i in range(39):
    dumb=pd.read_csv(f'.\\Data\\results\\30-08-2023, 12-22-11_heatflux_300_Sigmoid{i+1}.csv', index_col=0)
    profiles_sigmoid=np.vstack([profiles_sigmoid, dumb[dumb.columns[0]].values])

for i in range(30):
    dumb=pd.read_csv(f'.\\Data\\results\\24-07-2023, 21-53-28_heatflux_300_Sigmoid{40+1}.csv', index_col=0)
    profiles_sigmoid=np.vstack([profiles_sigmoid, dumb[dumb.columns[0]].values])

meanSigm=profiles_sigmoid.mean(axis=0)
stdSigm=profiles_sigmoid.std(axis=0)
```
